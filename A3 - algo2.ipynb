```python
import pandas as pd
import numpy as np
import random
import math
from copy import deepcopy
import random
import matplotlib.pyplot as plt
```

```python
def distance(a, b, ax=1):
    return np.linalg.norm(a - b, axis=ax)
```

```python
def algo2(data_frame, k):    
    C_index = random.sample(range(data_frame.shape[0]), 1)
    C = data_frame.values[C_index, :-1]
    for times in range(k-1):
        max_list = []
        for i in range(data_frame.shape[0]):
            distance(data_frame.values[i,:-1], C)
            max_list.append(np.min(distance(data_frame.values[i,:-1], C)))
        new_c = data_frame.values[np.argmax(max_list), :-1]
        C = np.vstack((C, new_c))

    clusters = {}
    for i in range(2):
        clusters[i] = []

    for i in range(data_frame.shape[0]):
        assigned_cluster = np.argmin(distance(data_frame.values[i,:-1], C))
        clusters[assigned_cluster].append(i)


    plt.figure(figsize=(10,10))
    temp = data_frame.iloc[clusters[0]]
    plt.scatter(temp.values[:,0], temp.values[:,1], c = 'b'  )
    temp = data_frame.iloc[clusters[1]]
    plt.scatter(temp.values[:,0], temp.values[:,1], c = 'r'  )
    plt.xlabel("1st - Dimension")
    plt.ylabel("2nd - Dimension")
    plt.title("Scatter plot for Algorithm - 2")


    plt.figure(figsize=(10,10))
    temp = data_frame[data_frame.c == 0]
    plt.scatter(temp.values[:,0], temp.values[:,1], c = 'b'  )
    temp = data_frame[data_frame.c == 1]
    plt.scatter(temp.values[:,0], temp.values[:,1], c = 'r'  )
    plt.xlabel("1st - Dimension")
    plt.ylabel("2nd - Dimension")
    plt.title("Scatter plot for Using label")
```

```python
df_1 = pd.read_csv('Dataset_1.csv', names = ['a','b','c'])
df_2 = pd.read_csv('Dataset_2.csv', names = ['a','b','c'])
df_3 = pd.read_csv('Dataset_3.csv', names = ['a','b','c'])
```

```python
algo2(df_1, 2)
algo2(df_2, 2)
algo2(df_3, 2)
```

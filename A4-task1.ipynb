```python
import pandas as pd
from sklearn import linear_model
import numpy as np
import matplotlib as plt
data=pd.read_csv('PB1_train.csv',sep=',',header=None)
data
```

```python
reg=linear_model.LinearRegression()
reg.fit(data[[0,1]],data[2])
```

```python
reg.coef_
```

```python
print('Model Parameters:-')
ϴ1= reg.coef_[0]
ϴ2= reg.coef_[1]
ϴ0=reg.intercept_
print('ϴ0=',ϴ0)
print('ϴ1=',ϴ1)
print('ϴ2=',ϴ2)
```

```python
print('The Given equation y=',ϴ0,'+',ϴ1,'x1''+',ϴ2,'x2')
```

```python
import pandas as pd
test_data=pd.read_csv('PB1_test.csv',sep=',',header=None,usecols=[0,1])
pred_data=reg.predict(test_data)
print('Predicted values on PB1_test.csv\n ',pred_data)
```

```python
given_data=pd.read_csv('PB1_test.csv',sep=',',header=None,usecols=[2])
from sklearn.metrics import mean_squared_error
m=mean_squared_error(pred_data,given_data)
print('Mean Squared Error=',m)
```

```python
p=reg.predict([[46,53]])
print('y value for x=[46,53] = ',p)
```

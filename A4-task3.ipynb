```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split 
from sklearn import metrics
```

```python
import numpy as np
import matplotlib as plt
data=pd.read_csv('PB3_train.csv',sep=',')
data.columns= ['HEIGHT','AGE','WEIGHT','LABEL']
data
```

```python
feature_cols = ['HEIGHT', 'AGE', 'WEIGHT']
x = data[feature_cols]
y = data.LABEL
```

```python
clf = DecisionTreeClassifier(criterion='gini')
clf=clf.fit(x,y)
```

```python
import pandas as pd
import numpy as np
import matplotlib as plt
test_data=pd.read_csv('PB3_test.csv',sep=',',usecols=[0,1,2])
test_data.columns= ['HEIGHT','AGE','WEIGHT']
test_data
```

```python
y_original=pd.read_csv('PB3_test.csv',sep=',',usecols=[3])
y_original.columns=['LABEL']
print('Original values on PB3_test.csv',y_original)
```

```python
y_pred = clf.predict(test_data)
print('Predicted values on PB3_test.csv',y_pred)
```

```python
print("Accuracy:",metrics.accuracy_score(y_original, y_pred)*100)
```
